```python
import pandas as pd
import sys
import os

sys.path.append('../')
from modules.data_cleaning import clean_text, normalize_text

print("[01_EDA_and_Cleaning]: Read raw data from: ../data/raw/train.csv")
df_train = pd.read_csv('../data/raw/train.csv')

df_train['comment_text'] = df_train['comment_text'].apply(lambda x: x[:2000])
df_train['clean_text_dl'] = df_train['comment_text'].apply(clean_text)
df_train['clean_text_ml'] = df_train['comment_text'].apply(normalize_text)

output_dir = '../data/interim/'
os.makedirs(output_dir, exist_ok=True)
output_file = os.path.join(output_dir, 'train_cleaned.csv')
df_train.to_csv(output_file, index=False)
print(f"[01_EDA_and_Cleaning]: Finish!")
```
