# DistilRoBERTa


```python
import pandas as pd
import numpy as np
import os
import torch
from sentence_transformers import SentenceTransformer

# 1. Kiểm tra lại "vũ khí"
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(f"Đang sử dụng thiết bị: {device}")
if device == 'cuda':
    print(f"Tên Card: {torch.cuda.get_device_name(0)}")

# 2. Tải mô hình Deep Learning (Nó sẽ tự tải weights về trong lần chạy đầu)
print("Đang load model all-distilroberta-v1...")
model = SentenceTransformer('all-distilroberta-v1', device=device)

# 3. Đọc dữ liệu đã clean từ bước trước
print("Đọc dữ liệu từ ../data/interim/train_cleaned.csv...")
df_train = pd.read_csv('../data/interim/train_cleaned.csv')

# Lấy cột text dành riêng cho DL và xử lý rỗng (phòng hờ)
texts = df_train['clean_text_dl'].fillna("").tolist()

# 4. TRÍCH XUẤT ĐẶC TRƯNG (Giai đoạn ngốn GPU nhất)
print("Bắt đầu trích xuất ma trận đặc trưng...")
# Với 16GB VRAM, batch_size = 128 hoặc 256 là mức lý tưởng để ép card chạy 100% công suất
X_train_embeddings = model.encode(
    texts, 
    batch_size=128, 
    show_progress_bar=True
)

# 5. Lưu ra file .npy theo chuẩn yêu cầu của môn học
save_dir = '../features/deep_learning/'
os.makedirs(save_dir, exist_ok=True)

output_path = os.path.join(save_dir, 'roberta_train_features.npy')
np.save(output_path, X_train_embeddings)

print(f"HOÀN TẤT! Đã lưu ma trận kích thước {X_train_embeddings.shape} tại: {output_path}")
```
