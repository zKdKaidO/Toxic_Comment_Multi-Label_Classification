```python
import pandas as pd
import numpy as np
import joblib
import time
from sklearn.linear_model import LogisticRegression
from sklearn.multiclass import OneVsRestClassifier

# 1. ĐỌC DỮ LIỆU ĐẦU VÀO (X và y)
print("1. Đang tải ma trận đặc trưng (X_train)...")
X_train = np.load('../features/deep_learning/roberta_train_features.npy')

print("2. Đang tải nhãn đáp án (y_train)...")
df_train = pd.read_csv('../data/interim/train_cleaned.csv')
label_cols = ["toxic", "severe_toxic", "obscene", "threat", "insult", "identity_hate"]
y_train = df_train[label_cols].values

# Kiểm tra khớp dữ liệu
assert X_train.shape[0] == y_train.shape[0], "Lỗi: Số dòng của X và y không khớp nhau!"

# 2. KHỞI TẠO MÔ HÌNH (Lắp não ra quyết định)
print("3. Đang thiết lập mô hình Logistic Regression (One-vs-Rest)...")
# Khởi tạo thuật toán lõi. 
# class_weight='balanced' chính là chìa khóa vàng để trị imbalanced data!
base_model = LogisticRegression(
    max_iter=1000,           # Cho phép thuật toán chạy nhiều vòng hơn để hội tụ
    class_weight='balanced', # Tự động phạt nặng nếu đoán sai comment Toxic
    solver='saga'            # Thuật toán tối ưu cực tốt cho ma trận lớn
)

# Bọc nó lại để xử lý 6 nhãn cùng lúc (Multi-label)
# n_jobs=-1 giúp huy động 100% sức mạnh của tất cả các nhân CPU trong máy bạn
classifier = OneVsRestClassifier(base_model, n_jobs=-1)

# 3. HUẤN LUYỆN (TRAINING)
print(f"4. BẮT ĐẦU TRAINING trên {X_train.shape[0]} mẫu dữ liệu...")
start_time = time.time()

# Dòng code quan trọng nhất của Machine Learning nằm ở đây:
classifier.fit(X_train, y_train)

# 4. LƯU THÀNH QUẢ
import os
save_dir = '../models/'
os.makedirs(save_dir, exist_ok=True)
model_path = os.path.join(save_dir, 'dl_roberta_logreg.joblib')

joblib.dump(classifier, model_path)

mins = (time.time() - start_time) / 60
print(f"HOÀN TẤT! Đã huấn luyện xong trong {mins:.2f} phút.")
print(f"Mô hình đã được lưu an toàn tại: {model_path}")
```

```python
import pandas as pd
import numpy as np
import joblib
import time
import os
from sklearn.linear_model import LogisticRegression
from sklearn.multiclass import OneVsRestClassifier

# 1. ĐỌC DỮ LIỆU DEBERTA
print("1. Đang tải ma trận đặc trưng DeBERTa (X_train)...")
X_train = np.load('../features/deep_learning/deberta_train_features.npy')

print("2. Đang tải nhãn đáp án (y_train)...")
df_train = pd.read_csv('../data/interim/train_cleaned.csv')
label_cols = ["toxic", "severe_toxic", "obscene", "threat", "insult", "identity_hate"]
y_train = df_train[label_cols].values

# Kiểm tra an toàn
assert X_train.shape[0] == y_train.shape[0], "Lỗi: Số lượng X và y không khớp!"

# 2. KHỞI TẠO BÁC SĨ (Giữ nguyên chiến thuật phạt nặng 'balanced')
print("3. Đang thiết lập Bác sĩ Logistic Regression cho DeBERTa...")
base_model = LogisticRegression(
    max_iter=1000,           
    class_weight='balanced', 
    solver='saga'            
)
classifier = OneVsRestClassifier(base_model, n_jobs=-1)

# 3. HUẤN LUYỆN
print(f"4. BẮT ĐẦU TRAINING trên {X_train.shape[0]} mẫu dữ liệu...")
start_time = time.time()

classifier.fit(X_train, y_train)

# 4. LƯU THÀNH QUẢ VỚI TÊN MỚI
save_dir = '../models/'
os.makedirs(save_dir, exist_ok=True)
# Đổi tên file để phân biệt với mô hình roberta cũ
model_path = os.path.join(save_dir, 'dl_deberta_logreg.joblib')

joblib.dump(classifier, model_path)

mins = (time.time() - start_time) / 60
print(f"HOÀN TẤT! Đã huấn luyện xong trong {mins:.2f} phút.")
print(f"Mô hình DeBERTa đã được lưu an toàn tại: {model_path}")
```
